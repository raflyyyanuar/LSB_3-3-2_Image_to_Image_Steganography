```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image and convert to numpy array
img = Image.open("image.jpg").convert('L')
img_arr = np.array(img)

# Flatten the image into a 1D array of pixel intensities
intensities = img_arr.flatten()

# Create a histogram of the pixel intensities
hist, bins = np.histogram(intensities, bins=256)

# Plot the histogram as a bar graph
plt.bar(bins[:-1], hist, width=1)

# Set labels and title
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.title('Grayscale Histogram')

# Show the plot
plt.show()
```
