## Import _library_

```python
from decimal import Decimal, getcontext
from collections import deque
from PIL import Image
import numpy as np
import math
import time
```

## Definisi fungsi

### Fungsi pembantu

```python
def get_image_size(image_path):
    size = Image.open(image_path).convert('RGB').size
    return (size[0] * size[1], size)
```

### Fungsi-fungsi steganografi

```python
# Fungsi untuk menyisipkan 1 byte ke 1 piksel
def embed_value(embed_value, pixel_value):
    # Pisah bit-bit yg sesuai dari 1 byte embed_value
    red_bits = (embed_value >> 5) & 0b111     # channel merah = 3 bit pertama
    green_bits = (embed_value >> 2) & 0b111   # channel hijau = 3 bit kedua
    blue_bits = embed_value & 0b11            # channel biru = 2 bit terakhir
    
    # Ganti bit-bit terakhir pada tiap channel piksel asli
    # dengan 1 byte yang telah dipisah
    new_red = (pixel_value[0] & ~0b111) | red_bits
    new_green = (pixel_value[1] & ~0b111) | green_bits
    new_blue = (pixel_value[2] & ~0b11) | blue_bits
    
    return new_red, new_green, new_blue
```

```python
# Fungsi utama untuk menyisipkan citra rahasia ke citra sampul
def embed_image(secret_path, cover_path, stego_path, locations, show_image=True):
    # Buka gambar dan ambil pikselnya
    secret_image = Image.open(secret_path).convert('RGB')
    secret_pixels = secret_image.load()
    
    cover_image = Image.open(cover_path).convert('RGB')
    cover_pixels = cover_image.load()
    
    # Ukuran minimal sampul adalah 3 kali ukuran citra rahasia
    if cover_image.width * cover_image.height < secret_image.width * secret_image.height * 3:
        raise ValueError("Cover must be three times secret image size.")
    
    # Penyisipan dimulai dari tiap piksel di citra rahasia
    # Piksel tersebut kemudian disebar ke tiga piksel di citra sampul
    
    # for loop di bawah dilakukan pada tiap piksel citra rahasia
    for i in range(len(locations)):
        if i >= secret_image.width * secret_image.height:
            break
        # for loop di bawah dilakukan pada tiap channel piksel citra sampul
        # c = 0 = merah
        # c = 1 = hijau
        # c = 2 = biru
        for c in range(3):
            # Tiap channel pada piksel citra rahasia disebar ke tiga piksel citra sampul 
            secret_rgb = secret_pixels[i % secret_image.width, i // secret_image.width]
            
            cover_pos = locations[3 * i + c]
            cover_pixel = cover_pixels[cover_pos % cover_image.width, cover_pos // cover_image.width]
            
            cover_pixels[cover_pos % cover_image.width, cover_pos // cover_image.width] = embed_value(secret_rgb[c], cover_pixel)
    
    # Menyimpan citra hasil penyisipan
    # Tunjukkan jika perlu
    cover_image.save(stego_path)
    if show_image:
        display(cover_image)
```

```python
# Fungsi utama untuk mengambil citra rahasia dari citra stego
def retrieve_image(size, stego_path, output_path, locations, show_image=True):
    # Buka gambar dan ambil pikselnya
    stego_image = Image.open(stego_path).convert('RGB')
    stego_pixels = stego_image.load()
    
    output_image = Image.new(mode="RGB", size=size)
    output_pixels = output_image.load()
    
    # for loop di bawah dilakukan pada tiap piksel citra output
    for i in range(len(locations)):
        if i >= size[0] * size[1]:
            break
        res = [0, 0, 0]
        
        # for loop di bawah dilakukan pada tiap channel piksel citra stego
        # c = 0 = merah
        # c = 1 = hijau
        # c = 2 = biru
        for c in range(3):
            stego_pos = locations[3 * i + c]
            r, g, b = stego_pixels[stego_pos % stego_image.width, stego_pos // stego_image.width]
            
            # Mengambil 3 bit terakhir channel merah, 3 bit terakhir channel hijau, dan 2 bit terakhir channel biru
            res[c] = int(f'{r & 0b111:03b}' + f'{g & 0b111:03b}' + f'{b & 0b11:02b}', 2)
        
        # Menaruh hasil pengambilan piksel asli ke citra output
        output_pixels[i % size[0], i // size[0]] = tuple(res)

    # Menyimpan citra hasil dekripsi
    # Tunjukkan jika perlu
    output_image.save(output_path)
    if show_image:
        display(output_image)
```

### Fungsi-fungsi manipulasi chaos

```python
# Fungsi utama untuk menghasilkan peta chaos sebagai persebaran piksel citra rahasia di citra sampul
def get_chaos(size, x0=Decimal('0.5'), r = Decimal('3.87'), i = 0, precision=16):
    # Penghitungan nilai chaos dilakukan menggunakan library Decimal
    # untuk mencapai presisi setinggi mungkin
    
    # Presisi desimal diatur oleh ini
    getcontext().prec = precision * 2
    
    # x0 dan r yang diberikan belum tentu bertipe Decimal
    # maka, konversi dilakukan
    x0 = Decimal(x0)
    r = Decimal(r)
    x = x0
    
    # untuk menghemat memori, generator digunakan sebagai pengganti array
    for _ in range(size):
        new_val = r * x * (1 - x)
        x = new_val

        # Hasil chaos dimodifikasi oleh ini
        # Modifikasi ini belum bisa melakukan persebaran merata
        # sehingga diperlukan pengujian lebih lanjut
        result = int(Decimal(new_val) * 4096)
        yield result
```

```python
# Fungsi utama untuk menghasilkan array yang menentukan persebaran piksel citra rahasia di citra sampul
def get_locations(secret_size, cover_size, x0=Decimal(0.5), precision=16):
    # Ukuran minimal sampul adalah 3 kali ukuran citra rahasia
    if cover_size < secret_size * 3:
        raise ValueError("Cover must be three times secret image size.")
    
    # Untuk mecapai performa tinggi,
    # tipe data deque akan digunakan sebagai pengganti array
    locations = deque([])
    
    # Area citra sampul disimpan di sini
    temp = deque(range(cover_size))
    
    # Hasil penghitungan peta chaos akan digunakan sebagai permutasi pada area citra sampul 
    chaos = deque(get_chaos(secret_size * 3, x0, precision=precision))
    chaos.reverse()
    
    # Untuk lebih menampakkan sifat chaotic,
    # permutasi Josephus dilakukan untuk menentukan persebaran piksel
    while len(chaos):
        pos = chaos.popleft()
        pos %= len(temp)
        
        # Dua line di bawah sepertinya masih bisa dioptimize lagi
        # karena "del" menggunakan O(n) dibanding "popleft" yg menggunakan O(1)
        # Untuk sejauh ini, belum ditemukan caranya
        locations.append(temp[pos])
        del temp[pos]
    
    return locations
```

### Fungsi analisis

```python
# Fungsi untuk menghitung MSE dan PSNR dua citra
def get_errors(a_path, b_path):
    # Buka citra
    a_image = Image.open(a_path).convert('RGB')
    b_image = Image.open(b_path).convert('RGB')
    
    # Mengubah citra ke array NumPy
    a_data = np.array(a_image)
    b_data = np.array(b_image)

    # Pastikan kedua citra seukuran
    if a_data.shape != b_data.shape:
        print(a_data.shape, b_data.shape)
        raise ValueError("Images must have the same dimensions.")
    
    # Menghitung Mean Squared Error (MSE)
    mse = np.mean((a_data - b_data) ** 2)
    
    # Menghitung Peak Signal-to-Noise Ratio (PSNR)
    max_pixel_value = 255  # Asumsi citra 8-bit
    psnr = 10 * math.log10((max_pixel_value ** 2) / mse)
    
    return (mse, psnr)
```
